```
import pandas as pd
IU = pd.read_csv('IU.csv',encoding='euc-kr',engine='python')
```

```
from konlpy.tag import Okt
from collections import Counter

okt=Okt()
```

```
IU=IU.drop('Unnamed: 0',axis=1)
```

```
IU
```

```
text=IU['1']
```

```
morphs=[]

for i in text:
  try:
    print(okt.pos(i))
    morphs.append(okt.pos(i))
  except:
    pass
```

```
noun_adj_adv_list=[]

for sentence in morphs:
  for word,tag in sentence:
    if tag in ['Noun']and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("게"not in word)and("말"not in word):
      noun_adj_adv_list.append(word)

print(noun_adj_adv_list)
```

```
for sentence in morphs:
  for word,tag in sentence:
    if tag in ['Adverb'] or ['Adjective']:
      noun_adj_adv_list.append(word)

print(noun_adj_adv_list)
```

```
count = Counter(noun_adj_adv_list)
print(count)
```

```
words=dict(count.most_common())
words
```
